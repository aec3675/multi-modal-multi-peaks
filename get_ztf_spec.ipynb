```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
header = ['Obj. ID', 'IAU name', 'Internal name/s', 'Obj. RA', 'Obj. DEC', 'Obj. Type',
          'Redshift', 'Spec. ID', 'Obs-date', 'JD', 'Phase (days)', 'From', 'Telescope',
          'Instrument', 'Observer/s', 'Reducer/s', 'Source group', 'Public', 'Associated groups',
          'End prop. period', 'Ascii file', 'Fits file', 'Spec. type', 'Spec. quality',
          'Extinction-Corrected', 'WL Medium',' WL Units', 'Flux Unit Coefficient', 'Spec. units',
          'Flux Calibrated By', 'Exp-time', 'Aperture (slit)', 'HA', 'Airmass', 'Dichroic', 'Grism',
          'Grating', 'Blaze', 'Lambda-min', 'Lambda-max', 'Del-Lambda', 'Contrib', 'Publish', 'Remarks',
          'Created by', 'Creation date']
all_spec = pd.read_csv('all_wiserep_spectra.csv', names=header, delimiter=', ')
```

```python
all_spec.columns
```

```python
all_spec['Obj. Type'].unique()
```

```python
# filters for specific types
multi_types = (all_spec['Obj. Type']=='SN IIb')|(all_spec['Obj. Type']=='SN Ib')|(all_spec['Obj. Type']=='SN Ic')|(all_spec['Obj. Type']=='SLSN-I')|(all_spec['Obj. Type']=='SLSN-II')|(all_spec['Obj. Type']=='SN IIn')

#filters for ZTF/P60 data
p60 = (all_spec['Telescope']=='P60')
sedm = (all_spec['Instrument']=='SEDM')
```

```python
mp_spec_df = all_spec.loc[multi_types&p60&sedm]
pd.to_numeric(mp_spec_df['Redshift'], errors='coerce')
mp_spec_df.shape[0]
```

```python
#filter based off specific type
iib = (mp_spec_df['Obj. Type']=='SN IIb')
ib = (mp_spec_df['Obj. Type']=='SN Ib')
ic = (mp_spec_df['Obj. Type']=='SN Ic')
slsni = (mp_spec_df['Obj. Type']=='SLSN-I')
slsnii = (mp_spec_df['Obj. Type']=='SLSN-II')
iin = (mp_spec_df['Obj. Type']=='SN IIn')

print(f"# of IIb: {mp_spec_df.loc[iib].shape[0]}")
print(f"# of Ib: {mp_spec_df.loc[ib].shape[0]}")
print(f"# of Ic: {mp_spec_df.loc[ic].shape[0]}")
print(f"# of SLSN-I: {mp_spec_df.loc[slsni].shape[0]}")
print(f"# of SLSN-II: {mp_spec_df.loc[slsnii].shape[0]}")
print(f"# of IIn: {mp_spec_df.loc[iin].shape[0]}")

#NOTE: probs lots of "misclassified" objs with un-updated types (e.g. 2021vgn, 2021gno, 2022hnt, ...) 
```
